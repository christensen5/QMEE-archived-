# Task 2

Plot the degree distribution of the air traffic network and fit an ER model. What is the maximum likelihood estimator for the connectivity parameter $p$? Can you derive an expression for the posterior if we take a Bayesian approach? Do you think an ER model would be able to capture the degree distribution? 

The maximum likelihood estimate of $p$ is simply the fraction of existing edges because the elements of the adjacency matrix are iid Bernoulli random variables.

```python
#Compute statistics
n = graph.number_of_nodes()
m = graph.number_of_edges()
eta = .5 * n * (n - 1)
#Compute the fraction
mle_p = graph.number_of_edges() / eta
print "Maximum likelihood estimate of p: {}".format(mle_p)
```

Taking a Bayesian approach, the likelihood of observing the entry $A_{ij}$ conditional on the parameter $p$ is
$$
P\left(A_{ij}|p\right) = p^{A_{ij}} (1-p)^{1-A_{ij}}.
$$
Because the elements of the adjacency matrix are independent the likelihood of the whole matrix is
$$
\begin{align}
P\left(A|p\right) &= \prod_{ij} p^{A_{ij}} (1-p)^{1-A_{ij}}\\
 &= p^m (1-p)^{\eta-m},
\end{align}
$$
where $\eta=n(n-1)/2$ is the number of edges if the network was fully connected. Using Bayes' theorem, the posterior is
$$
P\left(p|A\right)\propto P\left(A|p\right)P\left(p\right).
$$
We choose a Beta prior on $p$ with parameters $\alpha$ and $\beta$ because it is conjugate to the likelihood. Finally, the posterior is
$$
P\left(p|A\right)=\mathrm{Beta}\left(p;\alpha+m,\beta+\eta-m\right).
$$
We use Haldane's prior such that the mean of the posterior is equal to the maximum likelihood estimate.

```python
from scipy import stats

#Evaluate and plot the posterior
lin_p = np.linspace(0, .2, 100)
posterior_p = stats.beta.pdf(lin_p, m, eta-m)
plt.plot(lin_p, posterior_p, color='k')
#Plot the MLE estimate
plt.axvline(mle_p, color='k', linestyle='--')
plt.xlabel('$p$')
plt.ylabel('Posterior')
plt.tight_layout()
```

```python
#Get the degrees
degree = graph.degree().values()
#Get the frequency of degrees
frequency = np.bincount(degree)
frequency = 1.0 * frequency / np.sum(frequency)
#Plot the degree distribution
x = np.arange(frequency.shape[0])
plt.plot(x, frequency, color='k', marker='.', label='Degree sequence')

#Plot the MLE distribution
mle_dd = stats.binom.pmf(x, n - 1, mle_p)
plt.plot(x, mle_dd, color='r', marker='.', label='MLE distribution')

#Draw some samples from the beta distribution
samples_p = np.random.beta(m, eta-m, size=1000)
samples_dd = [stats.binom.pmf(x, n - 1, p) for p in samples_p]
plt.fill_between(x, np.percentile(samples_dd, 5, axis=0), np.percentile(samples_dd, 95, axis=0), 
                 color='r', label='5% - 95% credible interval', alpha=.2)

plt.xlabel('Degree')
plt.ylabel('Probability')
plt.legend(loc='upper right')
plt.tight_layout()
```

The degree sequence cannot be captured by an ER model.

# Task 5

This task is a little more involved: We will write functions to simulate a discrete-time random walk (DTRW) on a generic network.

Define a function `make_step` that takes two parameters: (1) a `graph` and (2) an `airport`. The function should return one of the neighbours of `airport` with probability proportional to the weight of the edge between `airport` and said neighbour. See http://docs.scipy.org/doc/numpy-dev/reference/generated/numpy.random.choice.html.

Define a function `make_step` that takes three parameters: (1) a `graph`, (2) an `airport`, and (3) `num_steps`. The function should return a sequence of length `num_steps` corresponding to the airports visisted by a random walker starting at `airport`. Make sure you reuse `make_step`.

```python
def make_step(graph, airport, weight='weight'):
    """
    Returns a neighbour of `airport` in `graph` with probability proportional to `weight`.
    """
    #Get the neighbours and the associated weight
    neighbours, weight = zip(*[(neighbour, 1.0 if weight is None else data['weight']) \
                               for neighbour, data in graph[airport].iteritems()])
    #Normalise the weights
    weight = np.asarray(weight) / np.sum(weight)
    #Choose a neighbour
    return np.random.choice(neighbours, p=weight)

def make_steps(graph, airport, num_steps, weight='weight'):
    """
    Returns a path of length `num_steps` starting frmo `airport` in `graph`. The probability 
    a neighbour is selected is probability proportional to `weight`.
    """
    path = [airport]
    #Iterate until we have made enough steps
    while len(path) <= num_steps:
        #Make a step...
        airport = make_step(graph, airport, weight)
        #...and append it to the path
        path.append(airport)
    return path

make_steps(graph, 'MEL', 4)
```

# Task 6

Construct the transition matrix $T_{ij}=\frac{A_{ij}}{k_j}$ and find its eigenvalues and eigenvectors. Find the steady-state solution of the DTRW on the air traffic network and visualise it. What is the characteristic timescale of the most slowly-decaying transient? How long should you run the simulation for to equilibrate the system?

```python
#Get the adjacency and transition matrix
adjacency = nx.adjacency_matrix(graph).toarray()
transition = adjacency / np.sum(adjacency, axis=0)

#Make sure the transition matrix is properly normalised
np.testing.assert_allclose(np.sum(transition, axis=0), 1)

#Compute the eigenvalues and eigenvectors
evals, evecs = np.linalg.eig(transition)
#Get a list of sorted indices
idx = np.argsort(np.abs(evals))[::-1]

#Make sure the largest eigenvalue is unity
np.testing.assert_almost_equal(evals[idx[0]], 1)

#Get the steady-state solution
steady_state = np.real(evecs[:, idx[0]])
steady_state = steady_state / np.sum(steady_state)

#Draw the coastlines
bm = Basemap()
bm.drawcoastlines()

#Draw the edges
nx.draw_networkx_edges(graph, pos, alpha=0.5)
#Draw the nodes
path_collection = nx.draw_networkx_nodes(graph, pos,
                                         node_color=steady_state, cmap='hot')
cb = plt.colorbar(path_collection)
cb.set_label('Steady-state probability')

#Focus on Australia
plt.xlim(110, 160)
plt.ylim(-45, -10)
plt.tight_layout()
```

```python
#Find the second largest eigenvalue in magnitude
lambda2 = np.real(evals[idx[1]])
timescale = -np.log(np.abs(lambda2))

print "Second-largest eigenvalue: {}".format(lambda2)
print "Characteristic timescale: {}".format(timescale)
```

The characteristic timescale is quite short so we only have to run the simulation for 5 or 10 steps.
